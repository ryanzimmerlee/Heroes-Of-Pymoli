## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Count the total number of unique items sold based on Item ID and place it in a variable 'items'
items = len(sales_df["Item ID"].unique())

# Find the average price of all of the items sold based on Price and place it in a variable 'avg_price'
avg_price = sales_df["Price"].mean()

# Find the number of unique purchases based on Purchase ID and place it in a variable 'purchases'
purchases = sales_df["Purchase ID"].count()

# Find the total revenue generated from the purchases based on Price and place it in a variable 'revenue'
revenue = sales_df["Price"].sum()

# Create a summary dataframe, labeling each variable created above appropriately
summary_data_df = pd.DataFrame({
    "Number of Unique Items Sold":[items],
    "Average Price":[avg_price],
    "Number of Unique Purchases":[purchases],
    "Total Revenue":[revenue]
})

# Format the Average Price and Total Revenue columns
summary_data_df["Average Price"] = summary_data_df["Average Price"].map('${:,.2f}'.format)
summary_data_df["Total Revenue"] = summary_data_df["Total Revenue"].map('${:,.2f}'.format)
summary_data_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed


```python
# Extract the columns SN, Age, and Gender for this exercise from the original dataframe that was imported
player_info = sales_df[["SN","Age","Gender"]]

# Drop the duplicates for each username as we only need one row for each username
player_info_uniq = player_info.drop_duplicates(subset=["SN"], keep = 'first')

# Found the total count of unique players/usernames
total_count = player_info_uniq["Gender"].count()

# Found the count of unique players by gender
gender_count = player_info_uniq["Gender"].value_counts()

# Create new dataframe to store gender count
summ_gender_count = pd.DataFrame({"Total Count":gender_count})

# Create new column within the new dataframe to calculate the percentage breakdown
summ_gender_count["Percentage of Players"] = gender_count/total_count*100

# Format for %
summ_gender_count["Percentage of Players"] = summ_gender_count["Percentage of Players"].map('{:,.2f}%'.format)

# Display the summary dataframe
summ_gender_count
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create data series for total count of purchases by Gender
gen_purc_count = sales_df['Gender'].value_counts()

# Create dataframe for total count of purchases by Gender
df_purc_count = pd.DataFrame({"Purchase Count":gen_purc_count})

# Create data series for mean of price by Gender
gen_mean = sales_df.groupby('Gender').mean()
gen_mean_pricecol = gen_mean.iloc[:,3]

# Create dataframe for mean of price by Gender
df_mean_pricecol = pd.DataFrame({"Average Price":gen_mean_pricecol})

# Format for $
df_mean_pricecol["Average Price"] = df_mean_pricecol["Average Price"].map('${:,.2f}'.format)

# Create data series for total of price by Gender
gen_total = sales_df.groupby('Gender').sum()
gen_total_pricecol = gen_total.iloc[:,3]

# Create dataframe for total of price by Gender
df_total_pricecol = pd.DataFrame({"Total Purchase Value":gen_total_pricecol})

# Format for $
df_total_pricecol["Total Purchase Value"] = df_total_pricecol["Total Purchase Value"].map('${:,.2f}'.format)

#Create data series for avg. total purchases per person by Gender
tot_pur_by_person = sales_df.groupby(['Gender','SN']).sum()
avg_pur_per_person = tot_pur_by_person.groupby('Gender').mean()
avg_pur_per_personcol = avg_pur_per_person.iloc[:,3]

# Create dataframe for avg. total purchases per person by Gender
df_avg_pur_per_personcol = pd.DataFrame({"Avg. Purchases per Person by Gender":avg_pur_per_personcol})

# Format for $
df_avg_pur_per_personcol["Avg. Purchases per Person by Gender"] = df_avg_pur_per_personcol["Avg. Purchases per Person by Gender"].map('${:,.2f}'.format)

#Concatenate all of the newly created dataframes...this works because index was maintained amongst all four dataframes
pd.concat([df_purc_count,df_mean_pricecol, df_total_pricecol, df_avg_pur_per_personcol ],axis=1,sort=True)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Pull from original dataframe the only columns needed
sub_df = sales_df[["Item ID","Item Name","Price"]]

# Group the above dataframe columns by Item ID and Item Name...do not sort...count the values
# This will provide 'Quantity Sold'
grsub_df = sub_df.groupby(['Item ID','Item Name'],sort=False).count()

# Group the above dataframe columns by Item ID and Item Name...do no sort...find avg. of the values
# This will provide 'Price per Item' 
item_price = sub_df.groupby(['Item ID','Item Name'],sort=False).mean()

# Grab the column with the 'Price per Item' from the dataframe and store it by itself for later
item_pricecol = item_price.iloc[:,0]

# Create new dataframe from the above data series we extracted
# This maintains the same index since we set sort=False
item_pricecol_df = pd.DataFrame({"Price per Item":item_pricecol})

# Create a summary dataframe of 'Quanity Sold' and 'Price per Item' by concatenating the two dataframes...
# Again, this works as Item ID and Item Name have same index due to NOT sorting
summ_table = pd.concat([grsub_df,item_pricecol_df],axis=1)

# Rename column to 'Quantity Sold' as this was not explicitly done above when counting
rename_summ_table = summ_table.rename(columns={"Price":"Quantity Sold"})

# Create the 'Total Purchase Value' column by multiplying the existing quantity and price
rename_summ_table["Total Purchase Value"] = rename_summ_table['Quantity Sold']*rename_summ_table['Price per Item']

# Sort the summary table based on 'Quantity Sold' column i.e. the PRODUCT that sold the most
sort_summ_table = rename_summ_table.sort_values(['Quantity Sold'],ascending=False)

# Display the summary table's top five rows
sort_summ_table.head()
```
